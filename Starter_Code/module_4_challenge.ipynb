## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df['subtotal'] = df['unit_price'] * df['qty']
df.iloc[0, :]
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
def shipping_rate(unit_weight):
    if unit_weight <51 :
        return unit_weight * 10
    else:
        return unit_weight * 7
    
df['shipping_price'] = df['unit_weight'].apply(shipping_rate)

df.iloc[0 , :]
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

# Create a function to calculate total price with a .0925 tax rate.

def total_price_calc(subtotal, shipping_rate):
    subtotal_price = (subtotal + shipping_rate)
    total_price = subtotal_price + (subtotal_price * .0925)
    return total_price

# Apply function to create a total_price column
df['total_price'] = df.apply(lambda row: \
total_price_calc(row['subtotal'], row['shipping_price']), axis = 1).round(2)

df.iloc[0 , :]
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
def line_cost_calc(unit_cost, qty, shipping_rate):
    total_cost = (unit_cost * qty) + shipping_rate
    return total_cost

df['line_cost'] = df.apply(lambda row: \
line_cost_calc(row['unit_cost'], row['qty'], row['shipping_price']),  axis = 1).round(2)

df.iloc[0, :]
```

```python
# Create a column for the profit of
# each line using line cost and line price
def profit_calc(line_cost, line_price):
    line_profit = line_price - line_cost
    return line_profit
    
df['bottom_line'] = df.apply(lambda row: \
profit_calc(row['line_cost'], row['subtotal']), axis = 1).round(2)

df.iloc[0, :]
```
